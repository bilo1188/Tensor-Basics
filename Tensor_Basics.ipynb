```
import torch
```

```
#1d
x = torch.empty(3)
print(x)
```

```
#2d
x = torch.empty(3,2)
print(x)
```

```
#3d
x = torch.empty(3,3,4)
print(x)
```

```
#4d
x = torch.empty(3,3,3,3)
print(x)
```

```
#2d
x = torch.zeros(2,2)
print(x)
```

```
#2d
x = torch.zeros(2,2,2,2)
print(x)
```

```
#2d
x = torch.ones(3,3)
print(x)
```

```
#2d
x = torch.ones(3,3,dtype =torch.float32)
print(x.dtype)
```

```
#2d
x = torch.ones(3,3,dtype =torch.float32)
print(x.size())
```

```
x= torch.tensor([2.5,4.5])
print(x)
```

```
x= torch.rand(2,2)
y= torch.rand(2,2)
print(x)
print(y)
print(x+y)
```

```
x= torch.rand(2,2)
y= torch.rand(2,2)
print(x)
print(y)
z = torch.add(x,y)
print(z)
```

```
x= torch.ones(2,2)
y= torch.ones(2,2)
#inplace addition
y.add_(x)
print(y)
```

```
x= torch.ones(2,2)
y= torch.ones(2,2)
#z = x-y
z = torch.sub(x,y)
print(z)
```

```
x= torch.ones(2,2)
y= torch.ones(2,2)
#z = x*y
z = torch.mul(x,y)
print(z)
y.mul_(x)
```

```
x= torch.ones(2,2)
y= torch.ones(2,2)
#z = x/y
z = torch.div(x,y)
print(z)
y.div_(x)
```
